```python
import matplotlib.pyplot as plt
import math
```

```python
MAX_N = 100
NUM_VALUES = 1000
x_values = [MAX_N/NUM_VALUES*n for n in range(NUM_VALUES)]
```

```python
fig, ax = plt.subplots()

y_values = x_values[:]
ax.plot(x_values, y_values, label=r'$\mathcal{O}(N)$')
y_values = [x**2 for x in x_values]
ax.plot(x_values, y_values, label=r'$\mathcal{O}(N^2)$')
y_values = [2**x for x in x_values]
ax.plot(x_values, y_values, label=r'$\mathcal{O}(2^N)$')
y_values = [math.factorial(x//1) for x in x_values]
ax.plot(x_values, y_values, label=r'$\mathcal{O}(N!)$')
y_values = [math.sqrt(x) for x in x_values]
ax.plot(x_values, y_values, label=r'$\mathcal{O}(\sqrt{N})$')
y_values = [math.log2(x) for x in x_values[1:]]
ax.plot(x_values[1:], y_values, label=r'$\mathcal{O}(\log_2{N})$')

ax.legend(loc='best', framealpha=0.0)
plt.ylim(0, 10)
plt.xlim(0, 10)
plt.grid(1)
```

```python
plt.show()
```
